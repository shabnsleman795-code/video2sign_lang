```python
# -*- coding: utf-8 -*-
"""Demucs on Colab: Separate audio tracks easily."""

# 1. ØªØ«Ø¨ÙŠØª Demucs ÙˆØ§Ù„Ù…ÙƒØªØ¨Ø§Øª Ø§Ù„Ø£Ø³Ø§Ø³ÙŠØ©
!python3 -m pip install -U demucs

# 2. Ø§Ø³ØªÙŠØ±Ø§Ø¯ Ø§Ù„Ù…ÙƒØªØ¨Ø§Øª
import os
import shutil
from pathlib import Path
import datetime

# 3. ØªØ­Ø¯ÙŠØ¯ Ù…Ø³Ø§Ø± Ø§Ù„Ù…Ù„Ù Ø§Ù„ØµÙˆØªÙŠ Ø£Ùˆ Ø§Ù„Ù…Ø±Ø¦ÙŠ ÙÙŠ Google Drive
input_file = "/content/drive/MyDrive/FinalVersion/Iameating.mp4"
input_dir = "/content/drive/MyDrive/FinalVersion/"
print(f"Using file from Google Drive: '{input_file}'")

# 4. Ø¥Ù†Ø´Ø§Ø¡ Ù…Ø¬Ù„Ø¯ Ø§Ù„Ù†ØªØ§Ø¦Ø¬ Ø¯Ø§Ø®Ù„ Ù†ÙØ³ Ø§Ù„Ø¯Ù„ÙŠÙ„
input_path = Path(input_file)
input_filename_without_ext = input_path.stem
model_name = "htdemucs"

# Ø¥Ù†Ø´Ø§Ø¡ Ù…Ø¬Ù„Ø¯ Results Ù…Ø¹ Ø·Ø§Ø¨Ø¹ Ø²Ù…Ù†ÙŠ
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
results_dir_name = f"Results_{input_filename_without_ext}_{timestamp}"
results_dir = Path(input_dir) / results_dir_name
results_dir.mkdir(exist_ok=True)

print(f"Creating results directory: '{results_dir}'")

# 5. ØªØ´ØºÙŠÙ„ Demucs Ù„ÙØµÙ„ Ø§Ù„Ù…Ø³Ø§Ø±Ø§Øª Ù…Ø¹ Ø­ÙØ¸ ÙÙŠ Ù…Ø¬Ù„Ø¯ Ù…Ø¤Ù‚Øª
temp_output_dir = "/content/temp_separated"

# Ø§Ø³ØªØ®Ø¯Ø§Ù… Ø§Ù„Ø£Ù…Ø± Ø§Ù„ØµØ­ÙŠØ­ Ø¨Ø¯ÙˆÙ† --dl
!demucs --two-stems vocals "{input_file}" -o "{temp_output_dir}"

# 6. Ø§Ù„Ø¨Ø­Ø« Ø¹Ù† Ø§Ù„Ù…Ù„ÙØ§Øª Ø§Ù„Ù…ÙØµÙˆÙ„Ø© ÙˆÙ†Ù‚Ù„Ù‡Ø§ Ø¥Ù„Ù‰ Ù…Ø¬Ù„Ø¯ Ø§Ù„Ù†ØªØ§Ø¦Ø¬
temp_sep_dir = Path(temp_output_dir) / model_name / input_filename_without_ext

if temp_sep_dir.exists():
    print(f"\nSeparation complete! Found results in: {temp_sep_dir}")

    # Ø¥Ù†Ø´Ø§Ø¡ Ù…Ø¬Ù„Ø¯Ø§Øª Ù…Ù†Ø¸Ù…Ø© Ø¯Ø§Ø®Ù„ Results
    vocals_dir = results_dir / "Vocals"
    accompaniment_dir = results_dir / "Accompaniment"
    vocals_dir.mkdir(exist_ok=True)
    accompaniment_dir.mkdir(exist_ok=True)

    # Ù†Ù‚Ù„ Ø§Ù„Ù…Ù„ÙØ§Øª Ø¥Ù„Ù‰ Ø§Ù„Ù…Ø¬Ù„Ø¯Ø§Øª Ø§Ù„Ù…Ù†Ø§Ø³Ø¨Ø©
    for file_path in temp_sep_dir.glob("*"):
        if file_path.is_file():
            file_name = file_path.name.lower()

            if "vocals" in file_name:
                dest_path = vocals_dir / file_path.name
                shutil.copy2(file_path, dest_path)
                print(f"Moved: {file_path.name} -> Vocals/")
            elif "other" in file_name or "no_vocals" in file_name:
                dest_path = accompaniment_dir / file_path.name
                shutil.copy2(file_path, dest_path)
                print(f"Moved: {file_path.name} -> Accompaniment/")
            else:
                dest_path = results_dir / file_path.name
                shutil.copy2(file_path, dest_path)
                print(f"Moved: {file_path.name} -> Root/")

    # 7. Ø¥Ù†Ø´Ø§Ø¡ Ù…Ù„Ù ØªÙ‚Ø±ÙŠØ±
    report_file = results_dir / "processing_report.txt"
    with open(report_file, 'w', encoding='utf-8') as f:
        f.write("Demucs Audio Separation Report\n")
        f.write("=" * 40 + "\n")
        f.write(f"Original file: {input_file}\n")
        f.write(f"Processing date: {datetime.datetime.now()}\n")
        f.write(f"Model used: {model_name}\n")
        f.write(f"Results directory: {results_dir}\n\n")

        f.write("Generated files:\n")
        for root, dirs, files in os.walk(results_dir):
            for file in files:
                if file != "processing_report.txt":
                    rel_path = os.path.relpath(os.path.join(root, file), results_dir)
                    f.write(f"- {rel_path}\n")

    print(f"\nAll files organized in: {results_dir}")
    print("Folder structure:")
    print(f"ðŸ“ {results_dir_name}/")
    print(f"   â”œâ”€â”€ ðŸ“ Vocals/")
    print(f"   â”œâ”€â”€ ðŸ“ Accompaniment/")
    print(f"   â””â”€â”€ ðŸ“„ processing_report.txt")

    # 8. ØªÙ†Ø¸ÙŠÙ Ø§Ù„Ù…Ù„ÙØ§Øª Ø§Ù„Ù…Ø¤Ù‚ØªØ©
    shutil.rmtree(temp_output_dir, ignore_errors=True)
    print("Temporary files cleaned up.")

else:
    print(f"\nSeparation failed. Temporary directory not found: {temp_sep_dir}")
    print("Checking for alternative output locations...")

    # Ø§Ù„Ø¨Ø­Ø« ÙÙŠ Ù…ÙˆØ§Ù‚Ø¹ Ø¨Ø¯ÙŠÙ„Ø©
    possible_locations = [
        temp_output_dir,
        "/content/separated",
        Path(temp_output_dir) / "htdemucs_ft",
        Path(temp_output_dir) / "mdx_extra_q"
    ]

    for location in possible_locations:
        loc_path = Path(location)
        if loc_path.exists():
            print(f"Found potential output at: {location}")
            # Ø³Ø±Ø¯ Ø§Ù„Ù…Ù„ÙØ§Øª Ø§Ù„Ù…ÙˆØ¬ÙˆØ¯Ø©
            for file_path in loc_path.rglob("*.*"):
                if file_path.is_file():
                    print(f"  - {file_path}")
```

```python
import json
import os
from pathlib import Path

def find_videos_by_extracted_words(json_file_path, video_directory):
    """
    Finds video files in a directory whose names contain words extracted from a JSON file.

    Args:
        json_file_path (str): The path to the JSON file containing the extracted words.
                              This file is expected to be a list of dictionaries,
                              where each dictionary has a 'text' key.
        video_directory (str): The path to the directory containing the video files.

    Returns:
        list: A list of paths to video files whose names contain any of the extracted words.
    """
    extracted_words = []
    try:
        with open(json_file_path, "r", encoding="utf-8") as f:
            word_matrix_from_json = json.load(f)

        # Extract words from the 'text' field of each dictionary in the list
        if isinstance(word_matrix_from_json, list):
            for item in word_matrix_from_json:
                if isinstance(item, dict) and 'text' in item:
                    # Split the text into words and add to the list
                    words_in_text = item['text'].split()
                    extracted_words.extend(words_in_text)
                elif isinstance(item, str):
                     # Handle cases where the list directly contains strings
                     words_in_text = item.split()
                     extracted_words.extend(words_in_text)


        # Remove duplicates and convert to lowercase for case-insensitive matching
        extracted_words = list(set([word.lower() for word in extracted_words if word]))

    except FileNotFoundError:
        print(f"Error: JSON file not found at {json_file_path}")
        return []
    except json.JSONDecodeError:
        print(f"Error: Could not decode JSON from {json_file_path}")
        return []
    except Exception as e:
        print(f"An unexpected error occurred while reading the JSON file: {e}")
        return []

    print(f"Extracted words for searching: {extracted_words}")

    found_videos = []
    if not Path(video_directory).is_dir():
        print(f"Error: Video directory not found or is not a directory: {video_directory}")
        return []

    print(f"Searching for videos in: {video_directory}")

    for filename in os.listdir(video_directory):
        file_path = Path(video_directory) / filename
        if file_path.is_file() and filename.lower().endswith(('.mp4', '.avi', '.mkv', '.mov', '.webm')): # Add other video extensions if needed
            file_name_lower = filename.lower()
            # Check if any extracted word is in the video filename
            if any(word in file_name_lower for word in extracted_words):
                found_videos.append(str(file_path))
                print(f"Found matching video: {filename}")

    return found_videos

# Example Usage (replace with your actual file paths)
json_file = "/content/words.txt" # Replace with the actual path to your JSON file
video_folder = "/content/drive/MyDrive/YourVideoFolder/" # Replace with the actual path to your video folder

matching_videos = find_videos_by_extracted_words(json_file, video_folder)

if matching_videos:
    print("\nMatching video files found:")
    for video in matching_videos:
        print(video)
else:
    print("\nNo matching video files found.")
```
